```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = [np.load("0_data.npy"), np.load("1_data.npy"), np.load("2_data.npy"), np.load("3_data.npy")]
```

```python
dflist = list()
cols = range(100,5001,100)
rows = [1] + list(range(5,101,5))
for k in data:
    newk = np.zeros((21,50))
    for i in range(21):
        for j in range(50):
            newk[i,j] =  k[i+j][3]
    dflist.append(pd.DataFrame(newk,columns=cols,index=rows))
```

```python
ax = sns.heatmap(dflist[0],cmap='viridis')
ax.set_xlabel("Iterations")
ax.set_ylabel("Alpha")
ax.set_title("8-3-8 Autoencoder \n Absolute Difference of Output from Input")
None
```
